```python
import pandas as pd
```

```python
df = pd.read_csv('../data/stocklists/gur.csv')
```

```python
print(df['segments_names'].value_counts())
```

```python
df_excluded = df[~df['segments_names'].isin(['Выводим из ассортимента'])]
print(df_excluded['segments_names'].value_counts())
```

```python
df = df_excluded.copy()
```

```python
df[['price_eur', 'price_usd']].hist(bins=20, figsize=(12, 5))
```

```python
print("price_eur:")
print(f"Zero: {(df['price_eur'] == 0).sum()}")
print(f"Non-zero: {(df['price_eur'] != 0).sum()}")
```

```python
print("price_usd:")
print(f"Zero: {(df['price_usd'] == 0).sum()}")
print(f"Non-zero: {(df['price_usd'] != 0).sum()}")
```

```python
df_non_0 = df[df['price_usd'] != 0]
df_non_0[['price_eur', 'price_usd']].hist(bins=20, figsize=(12, 5))
```
